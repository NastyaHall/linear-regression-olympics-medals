# Predicting amount of medals for teams
Here we will be using the teams dataset from 'teams.csv' file which can be found in the github repository by this link. 
Our goal is to build a model that will help us predict the amount of medals a team can possibly win based on the results of previous years. We are going to use scikit-learn's LinearRegression algorithms to build a model and make predictions.

## Extracting the data from csv

```python
import pandas as pd
```

```python
teams = pd.read_csv('teams.csv')
```

```python
teams
```

We have many different parameters such as team, country, year, events, etc. First we have to do a little analysis on our dataset to see if all the parameters are important, if there's any correlation between the medals (our target variable) and other parameters.

## Exploratory data analysis

```python
teams.describe()
```

The one thing worth noting is that columns such as prev_medals and prev_3_medals have less records (2014) than overall amount of records which is 2144. This means we will have to do some data cleaning like dropping records with missing values. Now let's see if there any correlations between the variables. We will visualise the data using seaborn library

```python
import seaborn as sns
sns.pairplot(teams, kind='scatter', plot_kws={'alpha':0.4}, diag_kws={'alpha':0.55, 'bins':40})
```

We can see a very clear correlation between medals and athletes and also between medals and prev_medals: the more atheletes there were, the more medals a team won; the more medals a team won in previous competition the more medals they won the next competition. What about the other variables? Well, for example athletes and events, height and weight make an extremely clear correlation but in our case it's not very important since we're trying to predict the amount of medals. Let's take a look at the numeral representation of the correlations between the variables to make sure our assumptions are right:

```python
teams_float = teams[['year', 'athletes', 'age', 'prev_medals', 'medals']]
teams_float.corr()['medals']
```

As we said amount of athletes and prev_medals have a great impact on the amount of medals team can possibly win: the closer the value is to 1, the more important it is. Something like year has a negative value at all, meaning there's absolutely no correlation between year and medals.

## Cleaning the data
Now it's time to remove unnecessary data to make it a little easier to work with the dataset. All we will do is drop the records that have missing values and get rid of columns that don't have much value in our situation.

```python
# seeing if there are missing values (i.e, Alb didn't have any medals in previous competition, because it didn't participate in it)
teams[teams.isnull().any(axis=1)]
```

As we assumed earlier prev_medals column had some missing values (130 records had no prev_medals). So, we're going to delete those. Also there's no need to work with the variables which don't play a significant role in prediciting the amount of medals and which don't display an important information. We will leave columns team , country, year, athletes, age, prev_medals and medals of course.

```python
# dropping rows with missing values
teams = teams[['team', 'country', 'year', 'athletes', 'age', 'prev_medals', 'medals']]
teams = teams.dropna()
teams
```

## Training the model 
Okay, we cleaned our data so now we can move to training the model. In this section we will split the data into train dataset and test dataset; train our model with train dataset and later on we will test it using the test dataset. The logic behind splitting the data in our specific situation is that we are training this model to predict how many medals a team can win in the future, so we don't necessarily need to use random_state, we can just take for train data years before 2012, and for test data years after 2012, as though we will be predicting the future.

```python
train = teams[teams['year'] < 2012].copy()
test = teams[teams['year'] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ['athletes', 'prev_medals']
target = 'medals'
```

We are using only athletes and prev_medals variables, since as we saw earlier in our little research, they have the largest impact on the amount of won medals unlike other parameters.

```python
reg.fit(train[predictors], train['medals'])
```

```python
predictions = reg.predict(test[predictors])
```

```python
test['predictions'] = predictions
test
```

We made our prediction, but it looks somewhat.. wrong. We are trying to predict the amount of medals, which can't be a fractional number as it is in the result. Also, we can see some values are negative which again is not acceptable. 
Let's do some corrections.

```python
# if the prediction is < 0, we simply round the prediction to 0
test.loc[test['predictions'] < 0, 'predictions'] = 0
```

```python
# here we round the fractional number to a whole integer
test['predictions'] = test['predictions'].round()
```

```python
test
```

Here we go, that looks much better, doesn't it? However, now we should evaluate our model. That's what we'll do in the next section

## Model evaluation
In this section we will see how well our model predicts the future by calculating the mean absolute error.

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test['medals'], test['predictions'])
error
```

The mean absolute error is ~ 3.3. What does this mean exactly? It simply means that on average we were within 3.3 medals of how many medals a team actually won in the Olympics. Now to find out whether or on it's a good value we need to dig a little bit deeper into our dataset. 

```python
teams.describe()['medals']
```

Let's look at the standard deviation: its value is 33.6. In a good scenario the mean absolute error should be below the standard deviation which is our case, 3.3 is far below 33.6. If it's higher than the standard deviation, that could mean that you used wronf predictors that just don't give you any information to predict the target or you messed something up in your model. Thankfully we are fine, let's make sure of it though. 

```python
test[test['team'] == 'USA'] 
```

So in 2012 USA won 248 medals and our model predicted 285, which is not a bad result at all. Same with 2016: USA actually won 264 medalsm and our model predicted 236. We aren't super accurate, but at the same time we're not that far from real values.

```python
errors = (test['medals'] - test['predictions']).abs()
errors_by_team = errors.groupby(test['team']).mean()
errors_by_team
```

Here we can see mean error for each team. Overall not so bad, but let's keep going.

```python
import numpy as np
medals_by_team = test['medals'].groupby(test['team']).mean()
error_ratio = errors_by_team / medals_by_team
error_ratio = error_ratio[np.isfinite(error_ratio)]
error_ratio
```

```python
error_ratio.plot.hist()
```

We can see that in some cases out prediction was far off real values, in some cases our predictions were pretty good. Whether it's good or bad depends on your purpose. It works well for countries that tend to win a lot of medals and it doesn't work very well for countries that normally don't win anything. We can see that by sorting our values: 

```python
error_ratio.sort_values()
```
